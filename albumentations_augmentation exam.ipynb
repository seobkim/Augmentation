# albumentations exam

```python
image = cv2.imread("cat2.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


# albumentations lib 이용하여 이미지 augamentation
transform = A.Compose([
    A.VerticalFlip(p=1)
])

#이미지 변환
transimg = transform(image=image)

#딕셔너리에서 array만 추출
transimg = transimg["image"]
print(type(transimg))


plt.imshow(transimg)




cv2.imshow('win',transimg)

cv2.waitKey()               
cv2.destroyAllWindows() 
```
